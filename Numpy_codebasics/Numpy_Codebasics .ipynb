3 main benefits of numpy array over python list
1. It needs less memory
2. It is faster 
3. It is convinient

list takes 14 bytes.

Below digram we can observe that on left hand side, we have numpy array which points to continuos location of memories.

Now we will see , how numpy array is faster than python list

```python
size=10000000

l1=range(size) #python lists
l2=range(size)

a1=np.arange(size) #Numpy array
a2=np.arange(size)

# python list
start=time.time()
result=[(x+y) for x,y in zip(l1,l2)] # we are adding elements in l1 and l2
                                      # and storing it in result
print("python list took:",(time.time()-start)*1000,end=" miliseconds") 
print()
#this will print time execution of addition of two pyton list
# and 1000 is multiplied to covert seconds to miliseconds

#numpy array
start=time.time()
result=a1+a2 # adding elements in 2 numpy arrays
print("numpy array took:",(time.time()-start)*1000,end=" miliseconds")
```

Itrate numpy array using nditer

```python
a=np.arange(12).reshape(3,4)
a
```

```python
#simple way of iterating is using for loop
for row in a:#we have seen this earlier
    for r in row:
        print(r)
```

```python
#by using flatten function we can do it in one for loop
for cell in a.flatten():
        print(cell)
```

```python
#numpy offers a funtion called nditer which allows more sofisticated way
#of iteration
for x in np.nditer(a,order='c'):#order c means elements will 
                                #get printed row by row
    print(x)
```

```python
for x in np.nditer(a,order='f'):#order f(fortran order) means elements will 
                                #get printed column by column
    print(x)
```

```python
#use of flags
#suppose we want to go in fortran order , but we dont want to print elements
#individually, we want columns to get printed, then we need to use flags
#for more details about flags in nditer see documentation by doing google
for x in np.nditer(a,order='f',flags=["external_loop"]):
        print(x)
```

```python
#while reading array we can modify it
for x in np.nditer(a,op_flags=["readwrite"]):
    x[...]=x*x
a
```

```python
#how to iterate two numpy arrays simultaneously?
b=np.arange(3,15,4).reshape(3,1)
b
```

```python
#we want to iterate through a and b simultaneously,but for this there is one 
#condition, that the dimensions of both matrices sholud be compatible
#( this is a genaral broadcasting rule)
for x,y in np.nditer([a,b]):
    print(x,y)
```
